```python
import pandas as pd
#set max columns
pd.set_option('display.max_columns', 52)

#reading in the complaints.csv data-set
complaints = pd.read_csv('service-requests.csv', dtype='unicode')
#only viewing the first 4 rows of the dataframe
complaints[:4]
```

```python
#head function displays the first 5 rows of the dataframe
complaints.head()
```

```python
#display a summary of the data-set
complaints.describe()
```

```python
#Display the first 20 rows but only of columns Complaint Type and Descriptor 
complaints[:20][['Complaint Type','Descriptor']]
```

```python
#Display how many types of unique complaints are there
complaints['Complaint Type'].nunique()
```

```python
#get a summary of the Complaint Type column
complaints['Complaint Type'].describe()
```

```python
#count the frequency of each Complaint Type and save the results in a new data-frame (complaint_counts)
complaint_counts = complaints['Complaint Type'].value_counts()
```

```python
#print complaint_counts data-frame
complaint_counts
```

```python
#Print the top ten rows
complaint_counts[:10]
```

```python
%matplotlib inline
complaint_counts[:10].plot(kind = 'bar', figsize = (15,10))
```

```python
#count the frequency of each occurence of a Borough in the the Borough column
complaints['Borough'].value_counts()
```

```python
complaints['Borough'].value_counts().idxmax
```

```python
#count the frequency of each occurence of a Borough in the Borough column and save it in a new dataframe
complaints_by_area = complaints['Borough'].value_counts()
```

```python
#plot the chart
complaints_by_area.plot(kind = 'bar', figsize=(10,5))
```

```python
#if the complaint type in the column is 'Noise - Street/Sidewalk', set the value to True else set it to False
is_noise = complaints['Complaint Type'] == "Noise - Street/Sidewalk"
#print is_noise
is_noise
```

```python
#for all the rows in complaints data-frame, if the 'is_noise' condition holds True, save that row in the noise_complaints dataframe
noise_complaints = complaints[is_noise]
noise_complaints
```

```python
#count the frequency of the occurence of each Borough
noise_complaints['Borough'].value_counts()
```

```python
#plot the graph
noise_plot = noise_complaints['Borough'].value_counts()
noise_plot.plot(kind = 'bar', figsize = (10,5))
```
